```python
import folium
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

data = pd.read_csv("../AB_NYC_2019.csv")
```

```python
data.head()
```

1. What is the variation range for property rental prices?

```python
mean = data.loc[:, "price"].mean()
sd = data.loc[:, "price"].std()
print(f"The average price is U${mean:.2f} +/- U${sd:.2f}")
```

2. What is the distribution of properties according to the minimum number of nights that can be booked?

```python
lines = data.loc[:, "minimum_nights"] < 100
data_plot = data.loc[lines, "minimum_nights"]
plt.hist(data_plot, bins=20)
```

3. What is the average and median number of properties registered per host?

```python
mean = data.loc[:, "calculated_host_listings_count"].mean()
median = data.loc[:, "calculated_host_listings_count"].median()
print(f"The average number of registered properties is {mean:.2f}")
print(f"AThe median number of registered properties is {median:.2f}")
```

5. How are the properties distributed among the host?

```python
data_plot = data["host_id"].value_counts()
print("Distribution of properties among hosts:")
print(data_plot)
plt.hist(data_plot, bins=100)
```

6. What is the most recent date of the last evaluation?

```python
reviews = data.loc[:, "last_review"].sort_values(ascending=True)
recent_review = reviews.loc[317]
print(f"The most recent date of evaluations is: {recent_review}")
```

7. What is the maximum value for scheduled overnight stays?

```python
max_overnight = data.loc[:, "minimum_nights"].max()
print(f"The maximum overnight stay is {max_overnight}")
```

9. How many hosts names are unique?

```python
hosts = data.loc[:, "host_name"].astype(str)
name_unique = len(np.unique(hosts))
print(f"The number of unique names of hosts is: {name_unique}")
```
